```python
def get_data(stk_num,path_in,date1,date2):
    if stk_num[:2]=='60':
        market='shanghai'
    elif stk_num[:2]=='00':
        market='shenzhen'
    else:
        print 'the stock number input is wrong'
        os._exit()
    stock_path1='\\'.join([path_in,market,stk_num])
    stk_list=os.listdir(stock_path1)
    num_in=find_stocks_in_date_range(stock_path1,date1,date2)
    #print len(num_in)
    price_all=[]
    vol_all=[]
    date_in=[]
    if len(num_in)>0:
        for fi1 in num_in:
            try:
                data_file='\\'.join([path_in,market,stk_num,stk_list[fi1]])
                #print data_file
                df1=read_vol_price_data(data_file)
                Count_Row=df1.shape[0]
                if Count_Row>2:
                    total1=df1.volume.values.tolist() ## pandas value to list
                    file_date=re.split('_|.csv',stk_list[fi1])[1]
                    #print total1[0]
                    price1=df1.price.values.tolist()
                    vol=[a/b for a, b in zip(total1,price1)]
                    price_all.append(price1)
                    vol_all.append(vol)
                    date_in.append(file_date)
            except:
                #print 'fuck'
                pass
    else:
        print 'no data'
    return price_all,vol_all,date_in
```

```python
stock_index_file="shanghai_shenzhen_data_from_2013"
stock_path='\\'.join([main_path_data,stock_index_file,market])
sto2=os.listdir(stock_path)
# define: date and market
date1=datetime(2017,1,16).strftime('%Y-%m-%d')
date2=datetime(2017,4,13).strftime('%Y-%m-%d')
#print date1+
stk_num="000099"
price_all,vol_all,date_in=get_data(stk_num,path_price_volume,date1,date2)
#print price_all[1]
print len(price_all)
print len(vol_all)
# price_all1 = list(itertools.chain(*price_all))
# vol_all1 = list(itertools.chain(*vol_all))
#np.max(vol_all1)

count_volume=[]
for cou in vol_all:
    count_volume.append(len(cou))
# print count_volume
# print date_in
df2=pd.DataFrame({'date':date_in,'vol':count_volume})
df2
```
