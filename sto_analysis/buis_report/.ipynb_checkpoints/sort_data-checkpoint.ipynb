```python
import os
import sys
#from load_path import *
import numpy as np
#import pandas as pd
import datetime
#from numpy._distributor_init import NUMPY_MKL
import re
from datetime import datetime
import itertools
from package_path_define.path_define import *
from package_readdata.read_data import * # load read data function
import pandas as pd

market='shanghai'
stock_index_file="shanghai_shenzhen_data_from_2013"
stock_path='\\'.join([main_path_data,stock_index_file,market])
sto2=os.listdir(stock_path)
dir1='business_report'
file_path='\\'.join([main_path_data,dir1,market])


```

```python
## col 0 每股收益
## col 1 每股收益扣除
## col 2 营业收入
## col 3 同比增长
## col 4 季度环比增长
## col 5 净利润
## col 6 净利润 同比增长
## col 8 每股净资产
## column 11 毛利润
# col 10每股经营现金流量
## col 13 股息率

ind=[]
stk_in=[]
for stk_num in sto2:
    file1=stk_num[0:6]+'.csv'
    data_file='\\'.join([file_path,file1])
    data1=pd.read_csv(data_file)
    #print data1.head()
    try:
        mean1=np.mean(data1.iloc[0:5,8])
    except:
        mean1=-999
    ind.append(mean1)
    stk_in.append(stk_num[0:6])
df2=pd.DataFrame({'stock':stk_in,'index':ind})
df2.sort_values(by='index',ascending = False)
#print data1.columns[6]
```
